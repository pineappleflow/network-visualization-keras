```python
from keras.applications.inception_v3 import InceptionV3
```

```python
inception = InceptionV3()
```

```python
def guess_seq_len(seq):
    guess = 1
    max_len = int(len(seq) / 2)
    for x in range(2, max_len):
        for y in range(2, max_len):            
            if seq[y:2*y] == seq[x:2*x]:
                print(seq[y:y*2])
                return x
    return guess
```

```python
layer_sequence = [type(layer) for layer in inception.layers]
```

```python
def get_repeated_seq(seq, start, length):
    ref = seq[start:start+length]

    for pos in range(start+length, len(seq)-length):
        compare = seq[pos:pos+length]
        
        if compare == ref:
            return pos
    return False

def get_repeated_seqs(seq):
    sequences = []
    for size in reversed(range(2, int(len(seq)/2))):
        for pos in range(0, len(seq)-size):
            sequences.append(get_repeated_seq(seq, pos, size))
    return sequences

print(get_repeated_seqs(layer_sequence))
```
